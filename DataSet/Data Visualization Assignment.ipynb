## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,100)
#dashdot or -. is use for marking the points with dashdot
plt.plot(x,np.sin(x),color='black',linestyle='dashdot')
#label is use for labelling
plt.xlabel('X',size=30,color='red')
plt.ylabel('Y',size=30,color='blue')
#title is use for giving the title to the graph
plt.title('Sine Wave',size=30,color='green')
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
S=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
C=np.array([82,78,48,46,39])
E=np.array([48,31,51,29,18])
A=np.array([15,20,14,13,9])
ind=[x for x, _ in enumerate(S)]
plt.bar(ind,C,width=0.8,label='Clothing',color='orange',bottom=E+A)
plt.bar(ind,E,width=0.8,label='Equipments',color='blue',bottom=A)
plt.bar(ind,A,width=0.8,label='Accessories',color='red')
plt.xticks(ind,S,rotation=90)
plt.xlabel("Cities")
plt.ylabel("Stores")
plt.legend(loc='upper right')
plt.title("Graph of different cities")
```

## Q.5 For Titanic Dataset

```python
data.head(7)
```
